# Questão 1

Implemente uma função load_data(seed) que:

Carregue o dataset `Fashion MNIST`
Realize a separação em treino e teste
Utilize `train_test_split` com controle de aleatoriedade
Retorne: `X_train`, `X_test`, `y_train`, `y_test`

Depois responda:
É necessário normalizar os dados para esse tipo de modelo? Justifique.

**Solução**:

````python
from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split

def load_data(seed=42):
    """
    Carrega o dataset Fashion MNIST, converte rótulos para inteiros
    e realiza a separação em treino e teste.
    """

    X, y = fetch_openml('Fashion-MNIST', version=1, as_frame=False, return_X_y=True, parser='auto')

    y = y.astype(int)

    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, random_state=seed
    )

    return X_train, X_test, y_train, y_test

#JUSTIFICATIVA
"""
Sim, a normalização é necessária (ou altamente recomendada) para a maioria dos modelos do sklearn.

Justificativa:
Os dados originais do Fashion MNIST consistem em pixels de 0 a 255.
1. Algoritmos baseados em gradiente (como Regressão Logística ou Redes Neurais/MLP)
   convergem muito mais rápido e são numericamente mais estáveis quando os dados
   estão em uma escala pequena (ex: 0 a 1).
2. Algoritmos baseados em distância (como SVM ou KNN) tratariam pixels com valores
   maiores como mais "importantes" se não houver normalização, distorcendo o aprendizado.
3. Modelos de árvores de decisão são invariantes à escala, mas como o objetivo é
   trabalhar com classificadores do sklearn de forma reprodutível e correta, a
   normalização (X / 255.0) garante consistência independentemente do modelo escolhido.
"""
````

# Questão 2

Implemente as funções:

`train_random_forest(X_train, y_train, seed)`
`train_adaboost(X_train, y_train, seed)`

## Requisitos:

Utilizar os modelos do `sklearn`
Garantir reprodutibilidade com `random_state`

**Solução**:

````python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

def train_random_forest(X_train, y_train, seed=42):
    """
    Treina um modelo Random Forest com controle de aleatoriedade.
    """
    rf_model = RandomForestClassifier(n_estimators=100, random_state=seed, n_jobs=-1)

    rf_model.fit(X_train, y_train)

    return rf_model

def train_adaboost(X_train, y_train, seed=42):
    """
    Treina um modelo AdaBoost com controle de aleatoriedade.
    """
    ada_model = AdaBoostClassifier(n_estimators=50, random_state=seed)

    ada_model.fit(X_train, y_train)

    return ada_model
````

# Questão 3

Implemente a função:

- `evaluate(model, X_test, y_test)`

Ela deve:
- Realizar predições
- Retornar a acurácia do modelo

**Solução**:

````python
from sklearn.metrics import accuracy_score

def evaluate(model, X_test, y_test):
    """
    Realiza predições em um conjunto de teste e retorna a acurácia.
    """
    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)

    return acc
````

# Questão 4

Implemente a função:

- `run_pipeline(model_type="rf", seed=42)`

Ela deve:
- Carregar os dados
- Treinar o modelo escolhido (`rf` ou `ab`)
- Avaliar o modelo
- Retornar a acurácia

**Solução**:

````python
def run_pipeline(model_type="rf", seed=42):
    """
    Executa o pipeline completo: carregamento, treinamento e avaliação.

    Parâmetros:
    model_type (str): "rf" para Random Forest ou "ab" para AdaBoost.
    seed (int): Semente para reprodutibilidade.
    """

    X_train, X_test, y_train, y_test = load_data(seed=seed)

    if model_type == "rf":
        model = train_random_forest(X_train, y_train, seed=seed)
    elif model_type == "ab":
        model = train_adaboost(X_train, y_train, seed=seed)
    else:
        raise ValueError("model_type deve ser 'rf' (Random Forest) ou 'ab' (AdaBoost).")

    accuracy = evaluate(model, X_test, y_test)

    return accuracy
````

Execute o pipeline para ambos os modelos:

- Random Forest
- AdaBoost

## Apresente:
- Acurácia, Precisão, Recall e F1-Score de cada modelo

## Responda:
- Qual modelo apresentou melhor desempenho inicial?

**Solução**:

````python
from sklearn.metrics import classification_report

def execute_comparison(seed=42):
    X_train, X_test, y_train, y_test = load_data(seed=seed)

    print("Treinando Random Forest...")
    rf_model = train_random_forest(X_train, y_train, seed=seed)
    y_pred_rf = rf_model.predict(X_test)
    report_rf = classification_report(y_test, y_pred_rf)

    print("Treinando AdaBoost...")
    ab_model = train_adaboost(X_train, y_train, seed=seed)
    y_pred_ab = ab_model.predict(X_test)
    report_ab = classification_report(y_test, y_pred_ab)

    print("\n" + "="*30)
    print("MÉTRICAS: RANDOM FOREST")
    print("="*30)
    print(report_rf)

    print("\n" + "="*30)
    print("MÉTRICAS: ADABOOST")
    print("="*30)
    print(report_ab)

execute_comparison(seed=42)

"""
Resposta:
O modelo que apresenta o melhor desempenho inicial no Fashion MNIST é o Random Forest.

Justificativa:
1. Complexidade dos Dados: O Fashion MNIST possui padrões espaciais complexos.
   O Random Forest, ao criar diversas árvores de decisão profundas, consegue capturar
   melhor as não-linearidades das imagens em comparação ao AdaBoost padrão.
2. Estimadores Base: O AdaBoost do sklearn utiliza, por padrão, 'Decision Stumps'
   (árvores com apenas 1 nível de profundidade), que são classificadores fracos demais
   para lidar com a alta dimensionalidade (784 features) deste dataset, resultando
   em métricas de Precision, Recall e F1-Score significativamente menores.
3. Robustez: O Random Forest é menos sensível a ruídos e tende a atingir uma
   acurácia acima de 85% rapidamente, enquanto o AdaBoost exige um ajuste fino
   muito maior ou estimadores base mais potentes para chegar perto desse resultado.
"""
````

Para pelo menos um dos modelos:

- Compare a acurácia em treino e teste

## Responda:
- Existe overfitting?
- Qual modelo tende a sofrer mais com isso?

````python
def check_overfitting(seed=42):
    """
    Compara a acurácia de treino e teste para o Random Forest.
    """
    X_train, X_test, y_train, y_test = load_data(seed=seed)

    rf_model = train_random_forest(X_train, y_train, seed=seed)

    acc_train = evaluate(rf_model, X_train, y_train)
    acc_test = evaluate(rf_model, X_test, y_test)

    print(f"Acurácia Treino: {acc_train:.4f}")
    print(f"Acurácia Teste: {acc_test:.4f}")

check_overfitting(seed=42)

"""
1. Existe overfitting?
Sim, há uma evidência clara de overfitting. No Fashion MNIST, é comum que o
Random Forest atinja uma acurácia próxima de 100% (ou 1.0) no conjunto de treino,
enquanto no conjunto de teste a acurácia costuma ficar entre 87% e 89%.
Essa discrepância (gap) indica que o modelo "memorizou" ruídos ou detalhes
específicos do treino que não generalizam perfeitamente para dados novos.

2. Qual modelo tende a sofrer mais com isso?
O Random Forest tende a sofrer mais com overfitting do que o AdaBoost padrão.

Justificativa:
- O Random Forest utiliza árvores de decisão profundas. Por padrão, o sklearn
  não limita a profundidade (max_depth=None), o que permite que cada árvore
  cresça até que todas as folhas fiquem puras, capturando cada detalhe do treino.
- O AdaBoost, por outro lado, utiliza "stumps" (árvores rasas de apenas 1 nível).
  Como cada estimador base é muito simples (um "underfitter" por natureza),
  é mais difícil para o ensemble como um todo memorizar o dataset de treino tão
  profundamente quanto uma floresta de árvores sem limites.
"""
````

# Questão 8

Varie pelo menos um hiperparâmetro em cada modelo:

- Random Forest: `n_estimators`
- AdaBoost: `n_estimators`

## Analise:
- O desempenho muda significativamente?

## Responda:
- Qual modelo é mais sensível a mudanças?

````python
def experiment_hyperparameters(seed=42):
    """
    Varia o n_estimators para Random Forest e AdaBoost e compara os resultados.
    """
    X_train, X_test, y_train, y_test = load_data(seed=seed)

    n_options = [10, 100]

    results = {"rf": {}, "ab": {}}

    for n in n_options:
        rf = RandomForestClassifier(n_estimators=n, random_state=seed, n_jobs=-1)
        rf.fit(X_train, y_train)
        results["rf"][n] = evaluate(rf, X_test, y_test)

        ab = AdaBoostClassifier(n_estimators=n, random_state=seed)
        ab.fit(X_train, y_train)
        results["ab"][n] = evaluate(ab, X_test, y_test)

    print(f"Resultados Random Forest: {results['rf']}")
    print(f"Resultados AdaBoost: {results['ab']}")

experiment_hyperparameters(seed=42)

"""
1. O desempenho muda significativamente?
No Random Forest, a mudança de 10 para 100 estimadores traz um ganho perceptível
de acurácia (geralmente de ~85% para ~88%), mas o modelo já começa com um
desempenho razoável mesmo com poucas árvores.
No AdaBoost, o impacto é visível, mas como o estimador base (stump) é muito
limitado, aumentar o número de estimadores melhora o modelo, mas ele ainda
tende a estagnar em um patamar de acurácia muito inferior ao Random Forest.

2. Qual modelo é mais sensível a mudanças?
O AdaBoost tende a ser mais sensível a mudanças nos hiperparâmetros.

Justificativa:
- O AdaBoost é um algoritmo de Boosting, onde cada árvore tenta corrigir o erro
  da anterior. Por ser um processo sequencial e dependente de 'weak learners',
  pequenas mudanças no número de estimadores ou na taxa de aprendizado (learning rate)
  podem alterar drasticamente a trajetória de convergência do modelo.
- O Random Forest é baseado em Bagging (processamento paralelo). Adicionar mais
  árvores geralmente apenas estabiliza a variância e melhora a acurácia de forma
  assintótica; ele é muito mais robusto e "funciona bem" logo de cara com as
  configurações padrão do sklearn, sendo menos propenso a falhar drasticamente
  por causa de um único hiperparâmetro.
"""
````
